# Correlating countries using pegasus for oppression to metrics such as press freedom index, democracy index etc.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('Data_On_Indices/Freedom_In_The_World/Ratings.xlsx', sheet_name = 'Country Ratings, Statuses ')
pegasusCountries = ['Bahrain', 'India', 'Mexico', 'Morocco', 'Saudi Arabia', 'United Arab Emirates'] # 'Palestine'

i = 1
CountryPrValues = []
avgOfCountryPrValues = []

# Collect political rights (PR) values for countries involved in pegasus scandal.
for index, row in df.iterrows():
    if row.array[0] in pegasusCountries:       
        for x in row.array:
            if(i < len(row.array)):
               CountryPrValues.append(row.array[i])
               i = i + 3
                
        avgOfCountryPrValues.append(list(CountryPrValues))
        CountryPrValues.clear()
        i = 1

#print('avgOfCountryPrValues = ' , avgOfCountryPrValues)
        
# Collect years to plot graph
yearIndex = 1
yearList = []

while True:
    if(i < len(df.iloc[0].array)):
        yearList.append(df.iloc[0].array[i])
        i = i + 3
    else:
        break
        
#print('yearList = ', yearList)

plt.figure(dpi=2000)

plt.plot(yearList, avgOfCountryPrValues[0], label=pegasusCountries[0])
plt.plot(yearList, avgOfCountryPrValues[1], label=pegasusCountries[1])
plt.plot(yearList, avgOfCountryPrValues[2], label=pegasusCountries[2])
plt.plot(yearList, avgOfCountryPrValues[3], label=pegasusCountries[3])
plt.plot(yearList, avgOfCountryPrValues[4], label=pegasusCountries[4])
plt.plot(yearList, avgOfCountryPrValues[5], label=pegasusCountries[5])

plt.title('Comparing the political right index of countries using pegasus.')
plt.xlabel('Years surveyed')
plt.ylabel('Political rights (Best - 1, Worst - 7)')
plt.xticks(np.arange(len(yearList)), yearList)
plt.legend(bbox_to_anchor=(1.05, 1))

plt.show()
```

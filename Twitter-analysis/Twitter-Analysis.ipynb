```python
# Reads the json generated from the CLI commands above and creates a pandas dataframe
import pandas as pd
import matplotlib.pyplot as plt

vijaita_tweets_df = pd.read_json('vijaita-query-tweets.json', lines=True)
sample = vijaita_tweets_df['date']
samplelist = list(sample)
sampledict = {}

for timestamp in samplelist:
    key = str(timestamp.month)
    if(sampledict.get(key) == None):
        sampledict[key] = 1
    else:
        sampledict[key] += 1
print(sampledict.keys())
print(sampledict.values())
plt.title('Number of tweets by Vijeta per month in 2021')
plt.xlabel('Months in 2021')
plt.ylabel('Number of tweets')
plt.bar(sampledict.keys(), sampledict.values())
```

```python
# Split dataset from digitalviolence.org into separate JSON files for loading into pandas
import json
with open('events.json', 'r') as f:
    digitalviolence_dataset = json.load(f)
    
with open('digitalviolence-events.json', 'w') as f:
    json.dump(digitalviolence_dataset['events'], f)
    
with open('digitalviolence-countries.json', 'w') as f:
    json.dump(digitalviolence_dataset['countries'], f)
    
with open('digitalviolence-targets.json', 'w') as f:
    json.dump(digitalviolence_dataset['targets'], f)
    
with open('digitalviolence-event_type_filters.json', 'w') as f:
    json.dump(digitalviolence_dataset['event_type_filters'], f)

with open('digitalviolence-categories.json', 'w') as f:
    json.dump(digitalviolence_dataset['categories'], f)

targets = set()
for row in digitalviolence_dataset['targets']:
    for entry in row.values():
        if isinstance(entry, str) and len(entry) != 0:
            targets.add(entry)
print('| --- | --- |')
print('| Name | Twitter |')
print('| --- | --- |')
for t in targets:
    print('|', t, '||')
```

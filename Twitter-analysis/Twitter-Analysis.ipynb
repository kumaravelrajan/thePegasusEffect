```python
import os

# Using OS library to call CLI commands in Python
os.system("snscrape --jsonl --since 2021-01-01 twitter-search \"from:vijaita until:2021-12-2\" > vijaita-query-tweets.json")
os.system("snscrape --jsonl --since 2021-01-01 twitter-search \"from:Khadija_Ismayil until:2021-12-2\" > Khadija_Ismayil-query-tweets.json")
os.system("snscrape --jsonl --since 2021-01-01 twitter-search \"from:aristeguicnn until:2021-12-2\" > aristeguicnn-query-tweets.json")
os.system("snscrape --jsonl --since 2021-01-01 twitter-search \"from:Nihalsingrathod until:2021-12-2\" > Nihalsingrathod-query-tweets.json")
os.system("snscrape --jsonl --since 2021-01-01 twitter-search \"from:Bagiruwubusa until:2021-12-2\" > Bagiruwubusa-query-tweets.json")
```

```python
# Reads the json generated from the CLI commands above and creates a pandas dataframe
import pandas as pd
import matplotlib.pyplot as plt

vijaita_tweets_df = pd.read_json('vijaita-query-tweets.json', lines=True)
sample = vijaita_tweets_df['date']
samplelist = list(sample)
sampledict = {}

for timestamp in samplelist:
    key = str(timestamp.month)
    if(sampledict.get(key) == None):
        sampledict[key] = 1
    else:
        sampledict[key] += 1
print(sampledict.keys())
print(sampledict.values())
plt.title('Number of tweets by Vijeta per month in 2021')
plt.xlabel('Months in 2021')
plt.ylabel('Number of tweets')
plt.bar(sampledict.keys(), sampledict.values())
```

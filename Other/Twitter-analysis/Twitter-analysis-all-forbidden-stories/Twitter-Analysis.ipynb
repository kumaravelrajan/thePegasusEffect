```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Using OS library to call CLI commands in Python

listOfJournos = ["ferdi_ayi", "khalafroula", "bradleyhope", "Khadija_Ismayil", "svaradarajan", 
                 "edwyplenel", "panyiszabolcs", "YuririaSierra", "Bagiruwubusa", "aristeguicnn", 
                 "vijaita", "paranjoygt", "aboubakrj", "LenaBred", "mkvenu1", "turankislakci", 
                 "Rosamoussaoui", "mariamoukrim", "jorgecarrascoa", "bainjal", "alexsicairosr", 
                 "Smita_Sharma", "xanic", "icembrero", "SushantSin", "MarocAmar", "marcelaturati", 
                 "carloskeiths", "ricardomraphael", "iftikhargilani", "JasurSumerinli", 
                 "rohini_sgh", "lhan55", "alvaro_delgado"]

#for journo in listOfJournos:
i = 1
for journo in listOfJournos:
    #os.system("snscrape --jsonl --since 2015-01-01 twitter-search \"from:" + journo + " until:2021-12-2\" > json-files/" + journo + "-query-tweets.json")
    
    # Reads the json generated from the CLI commands above and creates a pandas dataframe
    journo_tweets_df_old = pd.read_json("json-files/" + journo + '-query-tweets.json', lines=True)
    
    # Remove dates before Jan-2020
    journo_tweets_df_old['date'] = pd.to_datetime(journo_tweets_df_old['date'])
    journo_tweets_df = pd.DataFrame(journo_tweets_df_old[~(journo_tweets_df_old['date'] < '2020-01-01')])
    
    #display(journo_tweets_df)
    
    # Change date formats in df from "2021-12-01 18:00:00" -> 2021-December
    newDateFormat = {}.fromkeys(['date'], [])
    for timestamp in journo_tweets_df['date']:
        newDateFormat['date'].append(str(timestamp.year) + '-'+ timestamp.month_name())

    journo_tweets_df['date'] = newDateFormat['date']
    
    # Enumerate all unique YYYY-Month combinations.
    new_dates = journo_tweets_df.groupby(['date'], sort=False).groups.keys()
    list_new_dates = list(new_dates)
    list_new_dates.reverse()
    
    print(journo + ": " + str(list_new_dates))
    new_tweetsPerMonth = []
    
    # Calculate tweets per month
    for tweetsPerMonth in journo_tweets_df.groupby(['date'], sort=False).groups.values():
        new_tweetsPerMonth.append(len(tweetsPerMonth))
        
    new_tweetsPerMonth.reverse()
    print(str(new_tweetsPerMonth) + "\n\n\n")
    
    new_df = pd.DataFrame({'new_dates': list_new_dates,
                       'new_tweetsPerMonth': new_tweetsPerMonth})

    new_df.set_index(new_df['new_dates'], inplace=True)
    ax = new_df.plot.bar(x='new_dates', y='new_tweetsPerMonth', figsize=(20, 20), 
                    title="Tweets by " + journo + " since 2015")
    ax.figure.savefig("Graphs/" + str(i) + " - " + journo + ".png")
    plt.show()
    i += 1
```
